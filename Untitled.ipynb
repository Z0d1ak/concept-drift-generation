```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# set width of bar 
barWidth = 0.2
fig = plt.subplots(figsize =(12, 8)) 

# set height of bar 
# delta f1 pos | delta f1 negative | delta accuracy
CS = [0, -0.06, -0.04]
AS = [0.23, 0.03, 0.7]
LLM_AS = [0.33, 0.48, 0.48] 
LLM_SH = [0.09, 0.01, 0.3]  

# Set position of bar on X axis 
br1 = np.arange(len(CS)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 
br4 = [x + barWidth for x in br3] 

# Make the plot
plt.bar(br1, CS, color ='r', width = barWidth, 
        edgecolor ='grey', label ='Class shift') 
plt.bar(br2, AS, color ='g', width = barWidth, 
        edgecolor ='grey', label ='Adjective  swap') 
plt.bar(br3, LLM_AS, color ='b', width = barWidth, 
        edgecolor ='grey', label ='LLM assisted adjective swap')
plt.bar(br4, LLM_SH, color ='orange', width = barWidth, 
        edgecolor ='grey', label ='LLM assisted shortener') 

# Adding Xticks 
plt.xlabel('Metric', fontweight ='bold', fontsize = 12) 
plt.ylabel('Deviation from baseline', fontweight ='bold', fontsize = 12) 
plt.xticks([r + barWidth for r in range(len(CS))], 
        ['F1 Positive score', 'F1 Negative score', 'Accuracy'])

plt.legend()
plt.show() 
```
